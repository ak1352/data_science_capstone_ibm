# Scrape and Store Table into DataFrame

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df = pd.read_html(url)
df = df[0]
print(len(df))
df.head()
```

## Assigning Neighbourhood same as Borough if Neighbourhood is 'Not Assigned'

```python
for x in range(len(df)):
    if df['Borough'][x]!='Not assigned' and df['Neighbourhood'][x]=='Not assigned':
        df['Neighbourhood'][x] = df['Borough'][x]
df.head()
```

## Removing Rows with both Borough and Neighourhood as 'Not Assigned'

```python
df = df[(df['Borough'] != 'Not assigned')]
df.reset_index(drop = True)
len(df)
```

# Merge Neighbourhoods of Duplicate Borough

```python
df.groupby(['Postcode','Borough','Neighbourhood']).count()
```

```python
df=df.groupby("Postcode").agg(lambda x:','.join(set(x)))
```

```python
df.shape
```

```python
df_task_02 = pd.read_csv('Geospatial_Coordinates.csv')
df_task_02.head()
```

### Append Lat and Long to Original DataFrame

```python
for i in range(len(df)):
    if df.index[i] == df_task_02['Postal Code'][i]:
        df['Latitude'] = df_task_02['Latitude'][i]
        df['Longitude'] = df_task_02['Longitude'][i]
```

```python
df.head()
```

```python
df.to_csv('toronto_neighbourhood.csv')
```

### LAT AND LONG VALUE OF TORONTO 

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

### Visualize Toronto Area on Map

```python
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighbourhood']):
    label = '{}, {}'.format(df, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```
